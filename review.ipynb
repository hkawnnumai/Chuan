```python
str_given = "I love this movie! It is sweet, but with satirical humor. The dialogs are great and the adventure scenes are fun. It manages to be romantic and whimsical while laughingat the conventions of the fairy tale genre. I would recommend it to just about anyone. I have seen it several times and I am always happy to see it again and again. "
#Step 2: lower all the character to avoid case error
str_low = str_given.lower()

#Step 3: split the word to count
list = str_low.split()

#Step 4: to get back as a dictionary
counts = dict()

#Step 5: use for loop to count all the words in list
for word in list:
  
#Step 6: use if condition to check and count how many time each word appear
  if word in counts:
    counts[word] = counts[word] + 1
  else:
    counts[word] = 1

print (counts)

frequency = None
frequent_name = None
for key, value in counts.items():
  if frequency is None or value > frequency:
    frequency = value
    frequent_name = key
    
print("The most frequent is " ,frequent_name, " for ", frequency, "times.")


```
